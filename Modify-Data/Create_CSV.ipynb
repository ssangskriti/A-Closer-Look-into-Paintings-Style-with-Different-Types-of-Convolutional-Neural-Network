```python
mypath = "/home/protik/Downloads/Pandora_18k/Impressionism"
files = []
for (dirpath, dirnames, filenames) in walk(mypath):
    files.extend(filenames)
    break
```

```python
files
```

```python
print(len(files))
```

```python
row = []

for f in files:
    row.append(["impressionism",f])

with open('all_data_info.csv', 'a') as csvFile:
    
    writer = csv.writer(csvFile)
    writer.writerow(["genre","new_filename"])
    
    for i in range(0,len(row)):
        writer.writerow(row[i])

csvFile.close()
```
